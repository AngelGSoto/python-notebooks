```python
from __future__ import print_function
import numpy as np
from astropy.io import fits
import os
import glob
import json
import matplotlib.pyplot as plt
```

```python
datadir = "../../Obser-spectra-SOAR/2018-08-31/"
fitsfile = "wc0121_data_400m1-bg.fits"
hdulist= fits.open(os.path.join(datadir, fitsfile))
```

```python
hdulist.info()
```

```python
hdulist[0].header
```

```python
hdu = hdulist[0]
```

```python
len(hdu.data[0])
```

```python
Flux = hdu.data
Flux
```

```python
nx, wav0, i0, dwav = [hdu.header[k] for k in ("NAXIS1", "CRVAL1", "CRPIX1", "CD1_1")]
wavs = wav0 + (np.arange(nx) - (i0 - 1))*dwav 
```

```python
flux = hdu.data[1]
y = [y for y in flux]
hdu = hdulist[0]
len(y)
```

```python
plt.xlim(xmin=3000,xmax=7100)
#plt.ylim(ymin=-1e-17,ymax=1e-16)
xlabel("Wavelength, Angstrom")
ylabel("Flux, erg/cm^2/s/A")
plt.plot(wavs, flux)
gcf().set_size_inches(12,8)
grid()
plt.show()
plt.clf()
```
