```python
from __future__ import print_function
import numpy as np
import os
import glob
import json
import matplotlib.pyplot as plt
```

```python
datadir = "../PN-encontradas-azar-spectros/"
file_ = "k864.dat"
x = np.loadtxt(os.path.join(datadir, file_), delimiter = None, skiprows = 0, usecols = None, unpack = False, dtype = np.dtype([('Wl', '|f8'), ('Flux', 'f8')]))
```

```python
plt.xlim(xmin=3500,xmax=9900)
#plt.ylim(ymin=0.0,ymax=1.5e-13)
plt.xlabel('Wavelength $(\AA)$', fontsize=15)
plt.ylabel('Flux', fontsize=15)
plt.plot(x['Wl'], x['Flux'])
plt.savefig("../PN-encontradas-azar-spectros/spectrum-k864.jpg")
gcf().set_size_inches(12,8)
```
