# Find the allwise information of the known PN in VPHAS

```python
from astroquery.simbad import Simbad
import astropy.coordinates as coord
import astropy.units as u
from astropy.io import ascii
from astropy.table import Table
from astropy.coordinates import SkyCoord 
import numpy as np
from astroquery.vizier import Vizier
```

```python
# read the file
tab = Table.read("../../../PN_Giovanna_project/PN-HASH-zona2.dat", format="ascii")
tab
```

```python
result_vizier_wise = Vizier.query_region(coord.SkyCoord(ra=tab["RAJ2000_2"], dec=tab["DEJ2000"],
                                                        unit=(u.deg, u.deg),
                                                                       frame='icrs'),
                                                                           radius=4.0*u.arcsec,
                                                                              catalog=["allwise"])
```

```python
result_vizier_wise
```

```python
result_vizier_wise['II/328/allwise']
```

```python
# Save the table result WISE
asciifile2 = "../../../PN_Giovanna_project/PN-HASH-zona2-allwise.dat" 
result_vizier_wise['II/328/allwise'].write(asciifile2, format="ascii") 
```
