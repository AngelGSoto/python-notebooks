I made croos-match beteen a PNe sample from SMC (2007A&A...472..101I) and SMC catalog in SPLUS

```python
from astropy.io import fits
from pathlib import Path
import os
import numpy as np
import glob
import matplotlib.pyplot as plt
from astropy.wcs import WCS
from astropy.table import Table
#%matplotlib inline
%pylab inline

ROOT_PATH = Path("..")
```

```python
fitsfile = ROOT_PATH / "SMC/pne-smc-splus-match-2007.fits"
hdu = fits.open(fitsfile)
```

```python
tab = hdu[1].data
tab_ = Table(tab)
tab_
```

```python
#filters
r = tab["R_aper_3"]
f660 = tab["F660_aper_3"]
i = tab["I_aper_3"]
#colors
r_i = r - i
r_f660 = r - f660
```

```python
plt.xlabel('r - i', fontsize=15)
plt.ylabel('r - F660', fontsize=15)
plt.plot(r_i, r_f660, "o", label="Coll")
#plt.plot(x3['Wl'], x3['Flux'], label="15kpc")
#plt.plot(x4['Wl'], x4['Flux'], label="20kpc")
#plt.yscale("log")
#plt.legend()
#plt.savefig("spectrum-DdDm1.jpg")
gcf().set_size_inches(12,8)
```
