```python
from __future__ import print_function
import numpy as np
import os
import glob
import json
import matplotlib.pyplot as plt
```

```python
datadir = "../BB_DdDm1_L2-100-spectros/"
file_ = "DdDm1_L2_T50_output_SED.dat.tere_E0"
x = np.loadtxt(os.path.join(datadir, file_), delimiter = None, skiprows = 0, usecols = None, 
               unpack = False, dtype = np.dtype([('Wl', '|f8'), ('Flux', 'f8')]))
file_1 = "DdDm1_L2_T50_output_SED.dat.tere_E0.1"
x1 = np.loadtxt(os.path.join(datadir, file_1), delimiter = None, skiprows = 0, usecols = None, 
               unpack = False, dtype = np.dtype([('Wl', '|f8'), ('Flux', 'f8')]))
file_2 = "DdDm1_L2_T50_output_SED.dat.tere_E0.2"
x2 = np.loadtxt(os.path.join(datadir, file_2), delimiter = None, skiprows = 0, usecols = None, 
               unpack = False, dtype = np.dtype([('Wl', '|f8'), ('Flux', 'f8')]))
```

```python
#plt.xlim(xmin=3500,xmax=9900)
#plt.ylim(ymin=0.0,ymax=9e-13)
plt.xlabel('Wavelength $(\AA)$', fontsize=15)
plt.ylabel('Flux', fontsize=15)
plt.plot(x['Wl'], x['Flux'], label="E(B-V)=0.0")
plt.plot(x1['Wl'], x1['Flux'], label="E(B-V)=0.1")
plt.plot(x2['Wl'], x2['Flux'], label="E(B-V)=0.2")
plt.yscale("log")
plt.legend()
plt.savefig("spectrum-DdDm1.jpg")
gcf().set_size_inches(12,8)
```
