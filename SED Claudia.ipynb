```python
from __future__ import print_function
import numpy as np
from astropy.io import fits
import os
import glob
import json
import matplotlib.pyplot as plt
import sys
```

```python
datadir = "../Model_Claudia/"
file_ = "qso-z2p35.sed"
x = np.loadtxt(os.path.join(datadir, file_), delimiter = None, skiprows = 0, usecols = None, 
               unpack = False, dtype = np.dtype([('Wl', '|f8'), ('Flux', 'f8')]))
```

```python
plt.xlim(xmin=3000,xmax=9500)
#plt.ylim(ymin=0.0,ymax=9e-14)
plt.xlabel('Wavelength $(\AA)$', fontsize=15)
plt.ylabel('Flux', fontsize=15)
plt.plot(x['Wl'], x['Flux'])
plt.savefig("spectrum-DdDm1.jpg")
gcf().set_size_inches(10,7)
plt.show()
plt.clf()
```

```python
datadir = "../Model_Claudia/"
file_ = "star.sed"
x = np.loadtxt(os.path.join(datadir, file_), delimiter = None, skiprows = 0, usecols = None, unpack = False,
               dtype = np.dtype([('Wl', '|f8'), ('Flux', 'f8')]))
```

```python
asciifile = "star.dat"
file=open(asciifile,'w') #create file  
for x,y in zip(x['Wl'], x['Flux']):  
    file.write('%f  %s\n'%(x,y))     #assume you separate columns by tabs  
file.close()     #close file  
```

```python
plt.xlim(xmin=3000,xmax=9500)
plt.xlabel('Wavelength $(\AA)$', fontsize=15)
plt.ylabel('Flux', fontsize=15)
plt.plot(x['Wl'], x['Flux'])
plt.savefig("spectrum-DdDm1.jpg")
gcf().set_size_inches(10,7)
plt.show()
plt.clf()
```
