```python
%pylab inline
from __future__ import print_function
from astropy.io import fits
import os
import numpy as np
import glob
import matplotlib.pyplot as plt
from astropy.wcs import WCS
import sys
%matplotlib inline
```

```python
x0, x2 = [], []
datadir = "../SMC3/"
file_ = "19941017_ESO1.5m_SMC_3.dat"
x = np.loadtxt(os.path.join(datadir, file_), delimiter = None, skiprows = 0,
                usecols = None, unpack = False)
for a in x:
    x0.append(a[0])
    x2.append(a[1]*1.2e14)

```

```python
def load_file(filename):
    wll, ress = [], []
    data = np.loadtxt(filename, delimiter=None, converters=None, skiprows=0,
                                       usecols=None, unpack=False, ndmin=0)
    for i in data:
        wl = str(i[0])
        res = str(i[1])
        wll.append(wl)
        ress.append(res)
    return wll, ress
filename="../filters/VLT/SII2000trans.dat"
filename1="../filters/VLT/SII4500trans.dat"
plt.plot(x0, x2)
x, y = load_file(filename)
x1, y1 = load_file(filename1)
    
#plt.xlim(xmin=6600,xmax=7000)
plt.plot(x,y)
#plt.plot(x1,y1, linewidth=3.5)

plt.ylim(ymin=-0.0,ymax=0.82)
plt.xlabel('Wavelength $(\AA)$', fontsize=15)
#plt.ylabel('Flux', fontsize=15)

gcf().set_size_inches(12,8)
```
