```python
from __future__ import print_function
import numpy as np
from astropy.io import fits
import os
import glob
import json
import matplotlib.pyplot as plt
```

```python
datadir = "../ExtHIIr-NGC55/"
fitsfile = "M2S14b.0023.fits"
hdulist= fits.open(os.path.join(datadir, fitsfile))
fitsfile1 = "M2S14r.0023.fits"
hdulist1= fits.open(os.path.join(datadir, fitsfile1))
hdu = hdulist[0]
hdu1 = hdulist1[0]
nx, wav0, i0, dwav = [hdu.header[k] for k in ("NAXIS1", "CRVAL1", "CRPIX1", "CD1_1")]
wavs = wav0 + (np.arange(nx) - (i0 - 1))*dwav 
nx1, wav01, i01, dwav1 = [hdu1.header[k] for k in ("NAXIS1", "CRVAL1", "CRPIX1", "CD1_1")]
wavs1 = wav01 + (np.arange(nx1) - (i01 - 1))*dwav1 
Flux = hdulist[0].data
Flux1 = hdulist1[0].data
plt.xlim(xmin=3600,xmax=8000)
plt.ylim(ymin=-0.5e-17,ymax=0.3e-15)
xlabel("Wavelength, Angstrom")
ylabel("Flux, erg/cm2/s/A")
plt.plot(wavs, Flux)
plt.plot(wavs1, Flux1*1.25, color='red')
gcf().set_size_inches(12,8)
grid()
```
