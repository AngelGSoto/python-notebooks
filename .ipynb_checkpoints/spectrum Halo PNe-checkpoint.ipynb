```python
from __future__ import print_function
import numpy as np
import os
import glob
import json
import matplotlib.pyplot as plt
```

```python
datadir = "../Halo-PNe-spectros/"
file_ = "DdDm-1.dat"
x = np.loadtxt(os.path.join(datadir, file_), delimiter = None, skiprows = 0, usecols = None, unpack = False, dtype = np.dtype([('Wl', '|f8'), ('Flux', 'f8')]))
```

```python
plt.xlim(xmin=3500,xmax=9900)
plt.ylim(ymin=0.0,ymax=9e-14)
plt.xlabel('$Wavelength, Angstrom$')
plt.ylabel('Flux, erg/cm^2/s/Ang')
plt.plot(x['Wl'], x['Flux'])
#plt.savefig("spectro-DdDm1.jpg")
gcf().set_size_inches(12,8)
```
