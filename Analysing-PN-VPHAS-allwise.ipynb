# Plotting known PN VPHAS in allwise

```python
%pylab inline
```

```python
import numpy as np
import glob
import json
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sn
from astropy.table import Table
sn.set_context("poster")
```

```python
# read the file
tab = Table.read("../../PN_Giovanna_project/PN-HASH-zona2-allwise.dat", format="ascii")
tab.info
```

```python
#Defining the colours
w1_w4 = tab["W1mag"] - tab["W4mag"]
J_H = tab["Jmag"] - tab["Hmag"]
H_w2 = tab["Hmag"] - tab["W2mag"]
K_w3 = tab["Kmag"] - tab["W3mag"]
```

```python
#mask
m1 = w1_w4 >= 7.87 
m2 = J_H < 0.5
m3 = H_w2 >= 2.24
m4 = K_w3 >= 6.42
mask = m1 & m3 & m2 & m4
```

```python
fig, ax = plt.subplots(figsize=(12, 9))
ax.legend(ncol=2).set_title(r"Known PNe")
ax.set(xlim=[4.0, 13.0], ylim=[-0.5, 2.0])
#ax.set_aspect("equal")
ax.set(xlabel=r"$W1 - W4$", ylabel=r"$J - H$")
ax.scatter(w1_w4[mask], J_H[mask],  c=sn.xkcd_rgb['green'], 
                  alpha=0.9, marker='o', s=200, zorder=3, edgecolor='w', cmap="Blues")
#ax.plot(x, y, c="k", lw=0.9)
ax.axhline(y=1.10, c="k", lw=0.9)
ax.axvline(x=7.87, c="k", lw=0.9)
textbb = {"facecolor": "orange", "alpha": 0.7, "edgecolor": "none"}
textpars = {'ha': 'center', 'va': 'center', 'bbox': textbb, 'fontsize': 'small'}
ax.text(12, -0.19, r'PN zone', rotation=0, rotation_mode='anchor', **textpars)
```
