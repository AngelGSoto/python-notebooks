```python
from __future__ import print_function
import numpy as np
from astropy.io import fits
import os
import glob
import json
import matplotlib.pyplot as plt
```

```python
datadir = "../NGC55-HIIr/"
fitsfile = "H-10b.0016.fits"
hdulist= fits.open(os.path.join(datadir, fitsfile))
```

```python
datadir1 = "../NGC55-HIIr/"
fitsfile1 = "H-10r.0016.fits"
hdulist1= fits.open(os.path.join(datadir1, fitsfile1))
```

```python
hdulist.info()
```

```python
hdulist1[0].header
```

```python
hdu = hdulist[0]
hdu1 = hdulist1[0]
```

```python
hdu.data
```

```python
nx, wav0, i0, dwav = [hdu.header[k] for k in ("NAXIS1", "CRVAL1", "CRPIX1", "CD1_1")]
wavs = wav0 + (np.arange(nx) - (i0 - 1))*dwav 
nx1, wav01, i01, dwav1 = [hdu1.header[k] for k in ("NAXIS1", "CRVAL1", "CRPIX1", "CD1_1")]
wavs1 = wav01 + (np.arange(nx1) - (i01 - 1))*dwav1 
```

```python
Flux = hdulist[0].data
Flux1 = hdulist1[0].data
```

```python
plt.xlim(xmin=4090,xmax=5010)
plt.ylim(ymin=-1e-17,ymax=1e-16)
xlabel("Wavelength, Angstrom")
ylabel("Flux, erg/cm^2/s/A")
plt.plot(wavs, Flux)
plt.plot(wavs1, Flux1*0.60, color='red')
gcf().set_size_inches(12,8)
grid()
```
