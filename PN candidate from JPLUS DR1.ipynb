```python
from __future__ import print_function
import numpy as np
from astropy.io import fits
import os
import glob
import json
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline, BSpline

```

```python
datadir = "../spectrum-observed-INT/"
fitsfile = "Denise1_wbf-extr.fits"
hdulist= fits.open(os.path.join(datadir, fitsfile))
```

```python
hdulist.info()
```

```python
hdu = hdulist[0]
```

```python
hdu.data
```

```python
nx, wav0, i0, dwav = [hdulist[0].header[k] for k in ("NAXIS1", "CRVAL1", "CRPIX1", "CD1_1")]
wavelenght = wav0 + (np.arange(nx) - (i0 - 1))*dwav
wavelenght
```

```python
flux = np.array(hdu.data[1])
flux = flux.flatten()
flux
#x_smooth = np.linspace(wavelenght.min(), wavelenght.max(), 200)
#spl = make_interp_spline(np.array(wavelenght, dtype=str), power, k=3)
#y_smooth = BSpline(wavelenght, flux, k=3)
```

```python
canvas = plt.figure()
spl = canvas.add_subplot(1,1,1, axisbg='w')
plt.xlim(xmin=4000,xmax=7300)
plt.ylim(ymin=0.01e-18,ymax=0.4e-15)
xlabel("Wavelength, Angstrom")
ylabel("Flux, erg/cm^2/s/A")
spl.plot(wavelenght, flux)
#spl.spines['bottom'].set_color('r')
#spl.spines['top'].set_color('r')
#spl.spines['left'].set_color('r')
#spl.spines['right'].set_color('r')

gcf().set_size_inches(12,8)
grid()
plt.show()
plt.clf()
```
