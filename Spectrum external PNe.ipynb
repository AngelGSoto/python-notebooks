```python
from __future__ import print_function
import numpy as np
from astropy.io import fits
import os
import glob
import json
import matplotlib.pyplot as plt
```

```python
datadir = "../External-PN/"
fitsfile = "slit9b.fits"
hdulist= fits.open(os.path.join(datadir, fitsfile))
fitsfile1 = "slit9r.fits"
hdulist1= fits.open(os.path.join(datadir, fitsfile1))
hdu = hdulist[0]
hdu1 = hdulist1[0]
```

```python
nx, wav0, i0, dwav = [hdu.header[k] for k in ("NAXIS1", "CRVAL1", "CRPIX1", "CD1_1")]
wavs = wav0 + (np.arange(nx) - (i0 - 1))*dwav + 5.0
nx1, wav01, i01, dwav1 = [hdu1.header[k] for k in ("NAXIS1", "CRVAL1", "CRPIX1", "CD1_1")]
wavs1 = wav01 + (np.arange(nx1) - (i01 - 1))*dwav1 + 5.0
Flux = hdulist[0].data
Flux1 = hdulist1[0].data
#plt.xlim(xmin=6538,xmax=6600)
#plt.ylim(ymin=-1.9e-17, ymax=0.1e-15)
xlabel("Wavelength, Angstrom")
ylabel("Flux, erg/cm^2/s/A")
plt.plot(wavs, Flux+2.2e-18)
plt.plot(wavs1, Flux1+1.2e-18, color='red')
gcf().set_size_inches(12,8)
grid()
```

```python
datadir = "../External-PN/"
file_ = "slit9b+r.dat"
x = np.loadtxt(os.path.join(datadir, file_), delimiter = None, skiprows = 0, usecols = None, unpack = False, dtype = np.dtype([('Wl', '|f8'), ('Flux', 'f8')]))
#plt.xlim(xmin=6538,xmax=6600)
#plt.ylim(ymin=-1.9e-17, ymax=0.5e-15)
xlabel("Wavelength, Angstrom")
ylabel("Flux, erg/cm^2/s/A")
plt.plot(x['Wl'], x['Flux'], color='blue')
#plt.axvline(5700, color='k', linestyle='-', label="7082")
gcf().set_size_inches(12,8)
grid()
```
