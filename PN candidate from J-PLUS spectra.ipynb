```python
from __future__ import print_function
import numpy as np
from astropy.io import fits
import os
import glob
import json
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline, BSpline
```

```python
datadir = "../spectrum-observed-INT/Spectrum/"
file_ = "Denise1_wbf_ver1.ms.dat"
```

```python

x = np.loadtxt(os.path.join(datadir, file_), delimiter = None, skiprows = 0, usecols = None, unpack = False, dtype = np.dtype([('Wl', '|f8'), ('Flux', 'f8')]))
```

```python
plt.xlim(xmin=4000,xmax=7200)
plt.ylim(ymin=-0.1e-15,ymax=2.5e-15)
plt.xlabel('Wavelength $(\AA)$', fontsize=15)
plt.ylabel('Flux, $(\mathrm{erg s^{-1} cm^{-2} \AA^{-1}})$', fontsize=15)
plt.plot(x['Wl'], x['Flux'])
plt.savefig("../spectrum-observed-INT/Spectrum/Denise1_wbf-extr.pdf")
gcf().set_size_inches(12,8)
```
