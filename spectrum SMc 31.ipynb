```python
from __future__ import print_function
import numpy as np
import os
import glob
import json
import matplotlib.pyplot as plt
```

```python
datadir = "../SMC3/"
file_ = "19941017_ESO1.5m_SMC_3.dat"
x = np.loadtxt(os.path.join(datadir, file_), delimiter = None, skiprows = 0, usecols = None, unpack = False, dtype = np.dtype([('Wl', '|f8'), ('Flux', 'f8')]))
```

```python
fig = plt.figure(figsize=(12, 7))
ax1 = fig.add_subplot(111)
plt.xlim(xmin=4000,xmax=9000)
plt.ylim(ymin=0.0,ymax=0.8e-14)
plt.xlabel('Wavelength $(\AA)$', fontsize=15)
plt.ylabel('Flux (erg /cm$^2$ s $\AA$)', fontsize=15)
ax1.plot(x['Wl'], x['Flux'])
plt.savefig("spectrum-smc3.jpg")
gcf().set_size_inches(12,8)
```
