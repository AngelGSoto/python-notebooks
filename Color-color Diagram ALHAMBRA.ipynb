```python
def filter_mag(p, f1, f2, f3):
    col, col0 = [], []
    if data['id'].endswith(p) or data['id'].startswith(p):
        filter1 = data[f1]
        filter2 = data[f2]
        filter3 = data[f3]
        diff = filter1 - filter2
        diff0 = filter1 - filter3
        col.append(diff)
        col0.append(diff0)
    
    return col, col0
```
