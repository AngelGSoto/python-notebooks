# Finding for previosly publyshed information

```python
from astroquery.simbad import Simbad
import astropy.coordinates as coord
import astropy.units as u
from astropy.io import ascii
from astropy.table import Table
from astropy.coordinates import SkyCoord 
import numpy as np
from astroquery.vizier import Vizier
```

```python
# Read the table
tab = Table.read("../DR3-SPLUS/SySt-pc-halphaemitter-allmain-99-prob.tab", format="ascii.tab")
tab.info
```

```python
# Looking for the objects of tab in 2MASS through ViZier
for ra, dec in zip(tab["RA"], tab["DEC"]):
    result_vizier_2mass = Vizier.query_region(coord.SkyCoord(ra=ra, dec=dec,
                                                                  unit=(u.deg, u.deg),
                                                                                frame='icrs'),
                                                                                   radius=2.0*u.arcsec,
                                                                                         catalog=["2MASS"])
    result_vizier_2mass
    
```
